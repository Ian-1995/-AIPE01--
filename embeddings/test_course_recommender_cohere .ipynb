```python
import cohere
import pandas as pd

# 初始化 Cohere
co = cohere.Client("TbeJdxNDz62v8i0wYVE46BkyDBcphuSds92SMtpi")

# 準備 query 與文件
query = "WebXR（WebXR）:開發虛擬實境（VR）、擴增實境（AR）與混合實境（MR）體驗"
documents = course_df["課程相關總描述"].fillna("").tolist()

# 呼叫 Rerank 並要求回傳文件內容
response = co.rerank(
    model="rerank-multilingual-v3.0",
    query=query,
    documents=documents,
    top_n=20,
    return_documents=True,  # 才能取回 .document
)

# 顯示排序結果：分數、標題、網址、技能摘要
print("📊 Cohere 語意排序 Top 20 課程：\n")
for r in response.results:
    idx = r.index  # 取得排序對應原始資料 index
    title = course_df.iloc[idx]["課程名稱"]
    link = course_df.iloc[idx]["課程網址"]
    skill_desc = str(r.document).replace("  \n", " / ").strip()[:100]

    print(f"分數：{r.relevance_score:.3f}")
    print(f"課程名稱：{title}")
    print(f"課程網址：{link}")
    print(f"課程技能摘要：{skill_desc}")
    print("-" * 60)
```
