```python
import pandas as pd 
from pprint import pprint 

```

## 1️⃣ 查看資料基本結構與欄位型態

```python

df = pd.read_csv(r'104_programmers_jobs.csv')

df.info()
```

## 2️⃣ 顯示資料前幾列與後幾列

```python
display(df.head(3))
display(df.tail(3))
```

```python
df.duplicated().sum()
```

## 3️⃣ 檢查是否有重複的資料列

```python
missing_ratio = df.isnull().mean().sort_values(ascending=False)
display(missing_ratio)
```

## 4️⃣ 欄位遺失值統計

```python
missing_ratio = df.isnull().mean().sort_values(ascending=False)
display(missing_ratio)
```

## 5️⃣ 類別欄位與數值欄位分開列出

```python
cat_cols = df.select_dtypes(include="object").columns.tolist()
num_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
print("📌 類別欄位：", cat_cols)
print("📌 數值欄位：", num_cols)
```

## 6️⃣ 數值欄位描述統計

```python
df.describe()
```

## 7️⃣ 類別欄位唯一值統計

```python
for col in cat_cols:
    print(f"{col} 唯一值數量：{df[col].nunique()}")
```

## 8️⃣ 檢查類別欄位常見值與異常值

```python
for col in ['地區', '城市', '國家', '學歷要求', '工作經歷要求']:
    display(df[col].value_counts(dropna=False).head(10))
```

## 9️⃣ 檢查資料筆數是否有明顯異常（例如某地區大量缺值）

```python
df['地區'].isnull().groupby(df['城市']).sum().sort_values(ascending=False)
```
